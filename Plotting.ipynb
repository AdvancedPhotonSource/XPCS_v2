```python
import re

import h5py
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
f = h5py.File('./data/B010_Dragonite_25p_Quiescent_att2_Rq0_00001_0001_100000.hdf')
exchange_grp = [grp for grp in f.keys() if re.match('exchange', grp)]
print (exchange_grp)
for grp in exchange_grp:
    print(f[grp].keys())
    
frame_sum_orig = f['/exchange_orig/frameSum'][1:]
frame_sum_dev = f['/exchange/frameSum'][1:]

pixel_sum_orig = f['/exchange_orig/pixelSum'][1:]
pixel_sum_dev = f['/exchange/pixelSum'][1:]

partition_mean_total_orig = f['/exchange_orig/partition-mean-total'][0]
partition_mean_total_dev = f['/exchange/partition-mean-total'][0]

partition_mean_partial_orig = f['/exchange_orig/partition-mean-partial'][:]
partition_mean_partial_dev = f['/exchange/partition-mean-partial'][:]


f.close()
```

**Plots for Frames Sum**

```python
fig, ax = plt.subplots()

# ax.plot(frame_sum_orig[0])
# ax.plot(frame_sum_dev[0])

ax.plot(frame_sum_orig[0], label='Original')
ax.plot(frame_sum_dev[0], label='Dev')

legend = ax.legend(loc='upper center', shadow=True, fontsize='x-large')
```

```python
fig, ax = plt.subplots()

ax.plot(frame_sum_orig[0]-frame_sum_dev[0], label='Original')


frame_sum_dev[0][0]
```
